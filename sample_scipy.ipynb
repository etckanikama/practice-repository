```
from scipy.integrate import quad
import numpy as np
import matplotlib.pyplot as plt

def function(x):
  return x**2 + 3*x
# 引数：最小値、最大値、刻み
x = np.arange(-5,5,0.1)
# 目的関数、積分範囲で積分
I, _ = quad(function,0,3)
print(I)

y = function(x)

plt.plot(x,y)
plt.show()
```

```
# 三角関数のグラフを描く
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0,10,0.01)
y1 = sp.sin(x)
y2 = sp.cos(x)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y=sin(x) & y=cos(x)")
plt.plot(x,y1,label="sin")
plt.plot(x,y2,label="cos")
plt.legend()
plt.show()
```

```
# 統計学
import scipy as sp

# 合計
data = [i for i in range(0,10)]
print(sp.sum(data))
# 平均
print(sp.mean(data))
#分散
print(sp.var(data, ddof=0))
```

```
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt

x = np.array([0,1,2,3,4,5,6,6.5,7,7.5,8,8.5,9])
y = np.cos(-x**2/4.0)


f = interpolate.interp1d(x,y)

num = 50
xx = np.linspace(x[0],x[-1],num)
yy = f(xx)

plt.plot(x,y,label="cos")
plt.plot(xx,yy,label="gg")
plt.legend()
plt.show()
```
